```python
from pandas import *    
import seaborn as sbn

sbn.set(color_codes = True)

paths = []
paths.append('C:/Users/Fernando/Documents/Escritorio/excel files/order_products__prior.csv')
paths.append('C:/Users/Fernando/Documents/Escritorio/excel files/order_products__train.csv')
paths.append('C:/Users/Fernando/Documents/Escritorio/excel files/orders.csv')
paths.append('C:/Users/Fernando/Documents/Escritorio/excel files/products.csv')
```

```python
print("TAMAÑO DE DE DEPARTAMENTOS POR PRODUCTO")

import matplotlib.pyplot as plt

products = read_csv(paths[3])
depts_raw = products['department_id']
depts = []
depts_count = []
depts_labels = []


for dep in depts_raw:
    depts.append(dep)
    
for dep in depts:
    c = depts.count(dep)
    label = str(dep)
    depts_count.append(c)
    depts_labels.append(label)
    for i in range(c):
        depts.remove(dep)


colors = sbn.color_palette('pastel')[0:5]
plt.pie(depts_count, labels = depts_labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
print("CLIENTES QUE MAS COMPRAN")
orders = read_csv(paths[2])
users = orders['user_id']
users_  = []
for i in range(500):
    users_.append(users[i])
sbn.countplot(x=users_)
```

```python
print("PRODUCTOS MAS VENDIDOS")
orders_prior = read_csv(paths[0])
orders_train = read_csv(paths[1])
p_sold = orders_prior['product_id']
t_sold = orders_train['product_id']

products_sold = []

for i in range(250):
    products_sold.append(p_sold[i])
    products_sold.append(t_sold[i])

sbn.countplot(x=products_sold)
```
